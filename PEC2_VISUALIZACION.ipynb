```python
# Instalar libreras necesarias
!pip install matplotlib pandas

# Importar libreras
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame con tus datos
datos = pd.DataFrame({
    'Edades': ["Menos de 14 aos", "14 a 17 aos", "18 a 24 aos", "25 a 29 aos",
               "30 a 34 aos", "35 a 39 aos", "40 a 44 aos", "45 a 49 aos",
               "50 a 54 aos", "55 a 59 aos", "60 a 64 aos", "65 aos o ms"],
    'Mujeres': [86262, 26669, 37012, 32400, 39048, 37366, 35756, 35700, 33457, 33612, 31053, 84754],
    'Hombres': [89975, 27948, 39000, 34003, 39584, 37425, 36902, 36521, 33794, 33663, 30465, 73217]
})

# Convertir hombres a valores negativos
datos['Hombres'] = -datos['Hombres']

# Crear el grfico
fig, ax = plt.subplots(figsize=(10, 8))

# Barras horizontales
ax.barh(datos['Edades'], datos['Mujeres'], color='#FF69B4', label='Mujeres')
ax.barh(datos['Edades'], datos['Hombres'], color='#1E90FF', label='Hombres')

# Personalizar ejes
ax.set_xlabel('Poblacin')
ax.set_title('Pirmide poblacional Regin de OHiggins (2024)')
ax.legend()

# Mostrar valores absolutos en eje X
xticks = ax.get_xticks()
ax.set_xticklabels([abs(int(x)) for x in xticks])

# Mostrar grfico
plt.tight_layout()
plt.show()
```
